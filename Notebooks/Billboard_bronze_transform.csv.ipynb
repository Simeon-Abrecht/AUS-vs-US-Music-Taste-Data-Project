```python
import pandas as pd

# Path to your CSV file
file_path = "/Workspace/AUS vs US Music Taste/AUS-vs-US-Music-Taste-Data-Project/Data/Raw/Billboard/billboard_24years_lyrics_spotify.csv"

# Load CSV into a DataFrame
df = pd.read_csv(file_path)

# Peek at the first few rows
print(df.head())



```

```python
#inspect the headers of the dataframe
print(df.columns)
```

```python
#create a new dataframe with only the columns ranking, song, band_singer and year
df = df[['ranking', 'song', 'band_singer', 'year']]
#display the dataframe
display(df)
```

```python
df.columns = ["rank", "song_title", "artist", "year"]

display(df.head())
```

## creates a single artist column for every featuring artist, rather than splitting over mutiple rows 

```python
# Group by rank, title, and year; combine artists into a single string
df_combined = df.groupby(['rank', 'song_title', 'year'], as_index=False)\
                .agg({'artist': lambda x: ', '.join(x)})

df = df_combined

#revert to original ordering
df = df[['rank', 'song_title', 'artist', 'year']]
```

```python
display(df.head())
```

```python
# for loop to split dataframe based off of year column
for i in range(2000, 2024):
    df_year = df[df['year'] == i]
    df_year.to_csv(f'/Workspace/AUS vs US Music Taste/AUS-vs-US-Music-Taste-Data-Project/Data/Bronze/Billboard_bronze/billboard_{i}.csv', index=False)
```
